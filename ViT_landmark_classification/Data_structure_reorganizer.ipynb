# Reorgnaizing the files

```python
import numpy as np
import os
import pandas as pd
import cv2
from matplotlib import pyplot as plt
```

```python
old_path = 'C:/Adrian/Egyetem-Msc/Deep Learning Python es LUA alapon/Project/landmark-recognition-2021/'
train_data = pd.read_csv(old_path + 'train_corrected.csv')
train_data.index = train_data['Unnamed: 0'].values
train_data.drop('Unnamed: 0', axis='columns', inplace=True)
train_data
```

```python
train_data_50 = train_data[train_data['encoded_landmark_id'] < 50]
train_data_50
```

```python
# Creating new libraries


# Define the base directory where you want to create the libraries
new_path = "C:\Adrian\Egyetem-Msc\Deep Learning Python es LUA alapon\Project\TOP50_class_data"

# Create 50 libraries in the specified range
for i in range(50):
    library_name = str(i)
    library_path = os.path.join(new_path, library_name)

    # Check if the directory already exists
    if not os.path.exists(library_path):
        # Create the directory
        os.makedirs(library_path)
```

```python
# Now we grab all the pictures, and make a new home for them in a new directory

for ind, row in train_data_50.iterrows():

    # Reading the new picture
    image_id = row['id']
    file = image_id+'.jpg'
    subpath = '/'.join([char for char in image_id[0:3]]) 
    img = cv2.imread('C:\Adrian\Egyetem-Msc\Deep Learning Python es LUA alapon\Project\landmark-recognition-2021\\train/'+subpath+'/'+file)
    

    # Now we can drop it to its new folder
    output_path = os.path.join(new_path, str(row['encoded_landmark_id']), file)

    # Save the image as a JPEG file
    cv2.imwrite(output_path, img)
```
